# Relevant Data Overview Selection

```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
```

```python
dataset_file_name = "Data_Entry_2017.csv"
df = pd.read_csv(dataset_file_name)
```

```python
print(f"Dataframe size: {len(df)}")
```

```python
df.head(20)
```

## Filtering On Pneumonia And Age

```python
pneumonia_df = df[(df['Finding Labels'].str.contains('Pneumonia')) & (df['Patient Age'] >= 18)]
```

```python
print(f"Pneumonia dataframe size: {len(pneumonia_df)}")
```

```python
pneumonia_df.head(20)
```

## Filtering Equal Amount Of No Finding

```python
no_finding_df = df[(df['Finding Labels'] == 'No Finding') & (df['Patient Age'] >= 18)].head(len(pneumonia_df))
```

```python
print(f"No finding dataframe size: {len(no_finding_df)}")
```

## Combining And Exporting

```python
combined_df = pd.concat([pneumonia_df, no_finding_df])
```

```python
print(f"To be exported dataframe size: {len(combined_df)}")
```

```python
combined_df.to_csv("NIH_Filtered_Data_Entry.csv", index=False)
```
